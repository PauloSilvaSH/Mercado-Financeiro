```python
acoes = yf.download(tickers, start='2010-01-01', end=today)
print(acoes.head())
```

```python
acoes_tabular = acoes.stack(level=1).reset_index()

# Renomear colunas para facilitar entendimento
acoes_tabular.columns = ['Date', 'Ticker', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

acoes_tabular = acoes_tabular.rename(columns={
    'Ticker': 'Acoes',
    'Open': 'Abertura',
    'High': 'Maxima',
    'Low': 'Minima',
    'Close': 'Fechamento',
    'Adj Close': 'Fechamento Ajustado',
    'Volume': 'Volume (Quantidade)'
})
```

```python
tickers = acoes_tabular['Acoes'].unique()

# Inicializa DataFrame para armazenar todos os dividendos
dividendos_df = pd.DataFrame()

# Itera sobre os tickers e obtém dividendos
for ticker in tickers:
    stock = yf.Ticker(ticker)
    dividendos = stock.dividends
    dividendos = dividendos.reset_index()
    dividendos['Acoes'] = ticker  # Adiciona o ticker aos dividendos
    dividendos_df = pd.concat([dividendos_df, dividendos])  # Concatena os dividendos

# Renomear colunas do DataFrame de dividendos
dividendos_df.columns = ['Date', 'Dividendo', 'Acoes']

# Converte a coluna 'Date' para datetime, se necessário
acoes_tabular['Date'] = pd.to_datetime(acoes_tabular['Date'])
dividendos_df['Date'] = pd.to_datetime(dividendos_df['Date'])

# Faz o merge dos dividendos com o DataFrame 'acoes_tabular'
acoes_com_dividendos = pd.merge(acoes_tabular, dividendos_df, on=['Date', 'Acoes'], how='left')
```

```python
colunas_float = ['Abertura', 'Maxima', 'Minima', 'Fechamento', 'Fechamento Ajustado', 'Volume (Quantidade)','Dividendo']
```

```python
# Função para converter para string e substituir o ponto por vírgula
def formatar_decimal(valor):
    return str(valor).replace('.', ',') if valor is not None else ''

# Aplicar a formatação para cada coluna de float
for coluna in colunas_float:
    acoes_com_dividendos[coluna] = acoes_com_dividendos[coluna].apply(formatar_decimal)
```

```python
# Salvar o DataFrame em um arquivo CSV, usando vírgula como delimitador
acoes_com_dividendos.to_csv('acoes.csv', sep=';', index=False, decimal=',')
```
